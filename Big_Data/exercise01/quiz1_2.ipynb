## Q2

```python
%%sql
with release_durations as(
    select release_id, sum(cast(duration as float)) as duration
    from tracks
    group by release_id
)
select r.release_id, r.title, rd.duration
from released_by rb 
    join artists a using(artist_id) 
    join releases r using(release_id)
    join release_durations rd using(release_id)
where a.name = 'Radiohead'
order by rd.duration DESC, r.title
limit 1;
```

```python
%%sql
select t.release_id, r.title, sum(cast(t.duration as float)) as duration
from released_by rb
    join artists a using(artist_id) 
    join releases r using(release_id)
    join tracks t using(release_id)
where a.name = 'Radiohead'
group by t.release_id, r.title
order by duration DESC, r.title
limit 1;
```

## Q4

```python
%%sql
WITH releases_per_artist AS (
    SELECT artists.artist_id, artists.name, COUNT(*) AS release_count
    FROM artists
    JOIN released_by USING(artist_id)
    JOIN releases USING(release_id)
    GROUP BY artists.artist_id 
),
average_release_count AS (
    SELECT AVG(release_count)
    FROM releases_per_artist
),
electro_releases_per_artist AS(
    SELECT artists.artist_id, artists.name, COUNT(*) AS release_count
    FROM artists
    JOIN released_by USING(artist_id)
    JOIN releases USING(release_id)
    WHERE releases.genre = 'Electronic'
    GROUP BY artists.artist_id 
)
SELECT COUNT(*)
FROM electro_releases_per_artist
WHERE electro_releases_per_artist.release_count > (SELECT * FROM average_release_count);
```
